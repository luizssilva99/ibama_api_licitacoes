### Requisção a API (TESTE)

```python
import requests
import pandas as pd

# Configurações iniciais
tamanhoPagina = 10
orgao = "08829974000194"
anoPcaProjetoCompra = 2025
base_url = "https://dadosabertos.compras.gov.br/modulo-pgc/1_consultarPgcDetalhe"
headers = {"accept": "*/*"}

# Lista para armazenar os dados de todas as páginas
todos_dados = []

# Inicializa a iteração de páginas a partir da página 1
pagina = 1
while True:
    # Monta a URL com os parâmetros adequados
    url = f"{base_url}?pagina={pagina}&tamanhoPagina={tamanhoPagina}&orgao={orgao}&anoPcaProjetoCompra={anoPcaProjetoCompra}"
    
    # Realiza a requisição GET
    response = requests.get(url, headers=headers)
    response.raise_for_status()  # Levanta exceção para erros HTTP
    
    # Extrai os dados, considerando que eles estejam na chave "resultado"
    dados_pagina = response.json().get("resultado", [])
    
    # Se não houver dados, encerra a iteração
    if not dados_pagina:
        print(f"Encerrando a coleta: nenhum dado na página {pagina}")
        break
    
    # Adiciona os dados da página à lista geral
    todos_dados.extend(dados_pagina)
    print(f"Página {pagina} carregada com {len(dados_pagina)} registros.")
    
    # Incrementa a página para a próxima iteração
    pagina += 1

# Cria um DataFrame com os dados coletados e exibe as primeiras linhas
df = pd.DataFrame(todos_dados)
df.head()
```
